```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
ds=sns.load_dataset('iris')
```

```python
ds
```

```python
sns.lineplot(data=ds,color='black')
plt.show()
sns.boxplot(data=ds)
plt.show()
sns.violinplot(data=ds)
plt.show()
```

```python
g=sns.jointplot(data=ds,x="sepal_length",y="sepal_width",kind="reg",color="black")
g.plot_joint(sns.kdeplot)
```

```python
x=ds.iloc[:,0:5]
f,ax=plt.subplots(figsize=(5,6))
corr=x.corr()
print(corr)
sns.heatmap(corr,mask=np.zeros_like(corr),cmap=sns.diverging_palette(220,10, as_cmap=True),square=True,ax=ax,linewidth=.5)
plt.show()
```

```python
train,test=train_test_split(ds,test_size=0.40)
```

```python
train
```

```python
X_train=train[['sepal_length','sepal_width','petal_length','petal_width']]
y_train=train[['species']]
```

```python
test
```

```python
y_train
```

```python
X_test=test[['sepal_length','sepal_width','petal_length','petal_width']]
y_test=test[['species']]
```

```python
y_test
```

## SVM

```python
model=svm.SVC()
model.fit(X_train,y_train)
prd=model.predict(X_test)
prd
```

```python
metrics.accuracy_score(prd,y_test)
```

```python
print("Confusion Matrix\n",metrics.confusion_matrix(y_test,prd))
print("\n\nClassification Report",metrics.classification_report(y_test,prd))
print("Accuracy=",(metrics.accuracy_score(prd,y_test))*100,"%")
```

## K neighbours Classifier

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
```

```python
prd=model.predict(X_test)
```

```python
print("Confusion Matrix\n",metrics.confusion_matrix(y_test,prd))
print("\n\nClassification Report",metrics.classification_report(y_test,prd))
print("Accuracy=",(metrics.accuracy_score(prd,y_test))*100,"%")
```
